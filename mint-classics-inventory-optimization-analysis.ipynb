# 🚀 Inventory Optimization Analysis (Mint Classics Dataset)

## 📌 Project Overview
This project analyzes inventory data to identify inefficiencies and improve alignment between stock levels and sales performance.

The goal is to reduce overstocking, optimize warehouse distribution, and improve inventory decision-making.

## 🎯 Business Problem

- Inventory imbalance across products and warehouses  
- Overstocking leading to increased storage costs  
- Poor alignment between inventory levels and sales demand  
- Need to identify slow-moving vs high-performing products

## 🛠️ SQL Analysis

The following SQL queries were used to analyze inventory and sales data:

- Inventory overview  
- Sales performance  
- Inventory vs sales comparison  
- Inventory by category  
- Warehouse-level inventory distribution  

## 🔍 Key Insights

- Several products have high inventory but relatively low sales → overstocking risk  
- Weak correlation between inventory levels and demand  
- A small number of products drive the majority of sales  
- High-performing products are not always matched with sufficient inventory  
- Opportunity to rebalance stock and reduce holding costs  

## ✅ Conclusion

This analysis highlights key inefficiencies in inventory management and provides actionable insights to optimize stock distribution and reduce operational costs.
